```python
import pandas as pd
import numpy as np
import codecs

#with codecs.open('./gtop_smiles_2.csv', "r", "utf-8", "ignore") as file:
#    df = pd.read_table(file, delimiter=",",header=0)

df = pd.read_csv('./gtop_smiles_3.csv', header=0, encoding='cp932')

print(df)

```

```python
df2 = df.loc[:,["affinity_high","affinity_median","affinity_low"]]

#replace 0 with nan

df3 = df2.applymap(lambda x: np.nan if x == 0 else x)

#calc mean of affinity high and low, or use affinity median

df4 = df3.apply(np.mean, axis=1)

df5 = pd.Series(df4).to_frame()

df5.columns = ["affinity"]

df6 = pd.concat([df,df5],axis=1)

df6
```

```python
df7 = df6.loc[:,["target_gene_symbol","target_id","affinity","SMILES"]]

df7.dropna(subset=["SMILES"], inplace=True)

print(df7)
```

```python
import matplotlib.pyplot as plt

df7[df7.target_id == 21]


```

```python
c = df7["target_id"].as_matrix()
b = np.bincount(c)

print(c.max())

d = [i for i in range(c.max()+1)]
da = np.asarray(d, dtype=int)

e = np.stack((da.T,b.T), axis=1)

e = e[e[:,1].argsort()[::-1]]

np.savetxt('out.csv',e,delimiter=',')

plt.hist(c, bins=c.max()-1)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")


```
